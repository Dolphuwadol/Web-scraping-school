```python
total_pages = 236

# page= {} เพื่อค้นหาทุกหน้า
base_url = "http://www.thaischool.in.th/sitemap.php?page={}&school_area=&province_id=&txtsearch="

for page_number in range(1, total_pages + 1):
    url = base_url.format(page_number)
    print(url)
```

```python
# ข้อมูลทั้งหมดรวมใน list all_data
all_data = []

# page= {} เพื่อค้นหาทุกหน้า
base_url = "http://www.thaischool.in.th/sitemap.php?page={}&school_area=&province_id=&txtsearch="

for page_number in range(1, total_pages + 1):
    url = base_url.format(page_number)
    driver.get(url)
    data = driver.page_source
    df = pd.read_html(data)[2]
    df.columns = df.iloc[0]
    df = df[1:]
    df.reset_index(drop=True, inplace=True)

    # Append the data to the all_data list
    all_data.append(df)

# Combine all the DataFrames from each page into a single DataFrame
final_df = pd.concat(all_data, ignore_index=True)

# Print the final DataFrame
print(final_df)
```

```python
final_df
```

```python
school_data = "school_data.csv"
final_df.to_csv(school_data, index = False, encoding='utf-8-sig')
print("Data saved to CSV file:",school_data)
```
